Observable Trends
1. There is a relatively strong correlation between the latitude and temperature. The closer to zero the Latitude, the higher the temperature. This is consistent with the hypothesis that it gets hotter the closer you get to the equator.
2. There was little to no correlation between cloudiness and Latitude and Wind Speed and Latitude.
3. The correlation between humidity and Latitude is very weak.

```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import requests
import json
from citipy import citipy
import random
import csv 
#API key
from config import api_key
```

```python
#generate random coordinates
lat = []
lng = []

for x in range (2000):
    lat.append(random.randrange(-90,91,1))
for x in range (2000):
    lng.append(random.randrange(-180,181,1))
coordinates = list(zip(lat,lng))

#make sure they are unique
coordinates = list(set(coordinates))



print(len(coordinates))
    
```

```python
#create list of cities
cities = []

for coordinate_pair in coordinates:
    lat, lon = coordinate_pair
    cities.append(citipy.nearest_city(lat, lon).city_name)

countries = []
for coordinate_pair in coordinates:
    lat, lon = coordinate_pair
    countries.append(citipy.nearest_city(lat, lon).country_code)
City_Country = list(zip(cities,countries))
#make sure they are unique
City_Country = list(set(City_Country))

print(len(City_Country))
```

```python
#Break apart City_Country tuples so we have list of unique cities
city_list = [x[0] for x in City_Country]
country_list = [x[0] for x in City_Country]
```

```python
# API Info
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

query_url =  f"{url}appid={api_key}&units={units}&q="

```

```python
#Loop through and make API Request
weather_cities = []
new_city_list = []
new_country_List = []
Search_Counter = 0

for city in city_list:
    Search_Counter +=1
    response = requests.get(query_url + city).json()
    city_ID = response.get("id")
    if response.get("id"):
        weather_cities.append(response)
        new_city_list.append(city)
        new_country_List.append(city_list[Search_Counter-1])
    else:
        print(f"{city} not available" )


   
        
```

```python
len(weather_cities)
```

```python
#List comprehension
lat_data = [data.get("coord").get("lat") for data in weather_cities]
lng_data = [data.get("coord").get("lon") for data in weather_cities]
temp_data = [data.get("main").get("temp") for data in weather_cities]
humid_data = [data.get("main").get("humidity") for data in weather_cities]
cloud_data =[data.get("clouds").get("all") for data in weather_cities]
wind_data = [data.get("wind").get("speed") for data in weather_cities]
```

```python
new_cities_df = pd.DataFrame()
new_cities_series = pd.Series(new_city_list)
lat_data_series = pd.Series(lat_data)
lng_data_series = pd.Series(lng_data)
temp_data_series = pd.Series(temp_data)
humid_data_series = pd.Series(humid_data)
cloud_data_series = pd.Series(cloud_data)
wind_data_series = pd.Series(wind_data)


new_cities_df["City"] = new_cities_series.values
new_cities_df["Latitude"] = lat_data_series.values
new_cities_df["Longitude"] = lng_data_series.values
new_cities_df["Temperature"] = temp_data_series.values
new_cities_df["Humidity"] = humid_data_series.values
new_cities_df["Cloudiness"] = cloud_data_series.values
new_cities_df["Windiness"] = wind_data_series.values
new_cities_df.head()
```

```python
#create CSV
new_cities_df.to_csv("cities.csv", encoding="utf-8", index=False)
```

```python
#Temperature (F) vs Latitude
plt.figure(figsize=(20,10))
plt.scatter(new_cities_df["Latitude"], new_cities_df["Temperature"], s=200, color= "orange", edgecolor = "black", alpha =.75,)
plt.title('Temperature (F) vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.xlim(-90, 90)
plt.ylim(0,100)
plt.grid()
plt.savefig("Temp vs Lat.png")
plt.show()
```

```python
#Humidity (%) vs Latitude
plt.figure(figsize=(20,10))
plt.scatter(new_cities_df["Latitude"], new_cities_df["Humidity"], s=200, color= "c", edgecolor = "black", alpha =.75,)
plt.title('Humidity vs Latitude')
plt.xlabel('Latitude')
plt.ylabel('Humidity')
plt.xlim(-90, 90)
plt.ylim(0,110)
plt.grid()
plt.savefig("Humid vs Lat.png")
plt.show()
```

```python
#Cloudiness (%) vs Latitude
plt.figure(figsize=(20,10))
plt.scatter(new_cities_df["Latitude"], new_cities_df["Cloudiness"], s=200, color= "purple", edgecolor = "black", alpha =.75,)
plt.title('Cloudiness vs Latitude')
plt.xlabel('Cloudiness (%)')
plt.ylabel('Humidity')
plt.xlim(-90, 90)
plt.ylim(-10,110)
plt.grid()
plt.savefig("Cloud vs Lat.png")
plt.show()
```

```python
#Wind Speed (mph) vs Latitude
plt.figure(figsize=(20,10))
plt.scatter(new_cities_df["Latitude"], new_cities_df["Windiness"], s=200, color= "red", edgecolor = "black", alpha =.75,)
plt.title('Wind Speed (mph) vs Latitude')
plt.xlabel('Cloudiness (%)')
plt.ylabel('Wind Speed (mph)')
plt.xlim(-90, 90)
plt.ylim(-5,50)
plt.grid()
plt.savefig("Wind vs Lat.png")
plt.show()
```
